Cual es el precio medio de alquiler de cada uno, la diferencia que hay es estadsticamente significativa?
Con que test lo comprobaras?

```r
# Primero evaluamos con Test de ShapiroWilk para ver si proviene de una poblacin normalmente distribuida

for (tipo in levels(airbnb$TipoAlquiler)){df<-subset(airbnb,TipoAlquiler==tipo)
    print(paste("Precio medio tipo",tipo," = ",mean(df$Precio,na.rm=T)))
    if (nrow(df)>5000){df<-df[1:5000,]}
                                          
    sp<-shapiro.test(df$Precio)
                                          
    print(paste(" - Pvalor segn Test Shapiro-Wilk = ",sp$p.value))}
```

```r
#Segn los datos no hay una distribucin Gaussiana por lo que pruebo el test de Kruskal-Wallis
kruskal.test(Precio~TipoAlquiler,airbnb) -> kt
kt
kt$p.value                               -> Valor
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero son estadsticamente diferentes?
Que test habra que usar para comprobarlo?

```r
#Inicialmente valoramos el Test de Shapiro 
for (barrio in unique(pisos_grandes$Barrio)){
    pisos_grandes[which(pisos_grandes$Barrio==as.character(barrio)),]                -> df
    print(paste("Precio medio de",barrio," = ",mean(df$Metros.Cuadrados,na.rm=T)))
    if (nrow(df)>5000){
        df<-df[1:5000,]
    }
    shapiro.test(df$Metros.Cuadrados)                 -> sp
    print(paste(" - Test shapiro pvalor = ",sp$p.value))
}
# Al ser tan diferente entre s valoramos el test de Kruskal-Wallis
kruskal.test(Metros.Cuadrados~Barrio,pisos_grandes)    ->kt
kt
kt$p.value
paste("El valor del pvalor del test es muy alto") 
paste("Asumimos que el tamao de media de los pisos debera ser muy parecido y no tan distintos entre s")
```
