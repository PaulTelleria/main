Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

Cual es el precio medio de alquiler de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
# Primero evaluamos con Test de Shapiro–Wilk para ver si proviene de una población normalmente distribuida

for (tipo in levels(airbnb$TipoAlquiler)){df<-subset(airbnb,TipoAlquiler==tipo)
    print(paste("Precio medio tipo",tipo," = ",mean(df$Precio,na.rm=T)))
    if (nrow(df)>5000){df<-df[1:5000,]}
                                          
    sp<-shapiro.test(df$Precio)
                                          
    print(paste(" - Pvalor según Test Shapiro-Wilk = ",sp$p.value))}
```

```r
#Según los datos no hay una distribución Gaussiana por lo que pruebo el test de Kruskal-Wallis
kruskal.test(Precio~TipoAlquiler,airbnb) -> kt
kt
kt$p.value                               -> Valor
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
#Inicialmente valoramos el Test de Shapiro 
for (barrio in unique(pisos_grandes$Barrio)){
    pisos_grandes[which(pisos_grandes$Barrio==as.character(barrio)),]                -> df
    print(paste("Precio medio de",barrio," = ",mean(df$Metros.Cuadrados,na.rm=T)))
    if (nrow(df)>5000){
        df<-df[1:5000,]
    }
    shapiro.test(df$Metros.Cuadrados)                 -> sp
    print(paste(" - Test shapiro pvalor = ",sp$p.value))
}
# Al ser tan diferente entre sí valoramos el test de Kruskal-Wallis
kruskal.test(Metros.Cuadrados~Barrio,pisos_grandes)    ->kt
kt
kt$p.value
paste("El valor del pvalor del test es muy alto") 
paste("Asumimos que el tamaño de media de los pisos debería ser muy parecido y no tan distintos entre sí")
```

¿Que variable es más fiable para conocer el precio de un inmueble, el número de habitaciones o los metros cuadrados?

```r
#Pruebo las distintas posibilidades para ver qué variable influye en mayor medida en el precio
model<-lm(barrio_sol,formula=Precio~NumBanyos+NumDormitorios+Metros.Cuadrados+MaxOcupantes)
summary(model)
```

```r
#Primero la variable de Metros.Cuadrados
model1<-lm(barrio_sol,formula=Precio~Metros.Cuadrados)
summary(model1)
#Compruebo que el p-valor es 6x10^(-8)
```

```r
#Después la variable de NumDormitorios
model2<-lm(barrio_sol,formula=Precio~NumDormitorios)
summary(model2)
#Compruebo que el error es de 3.6 x 10^(-11) , aproximadamente la milésima parte con respecto del anterior
```

```r
#De manera que compruebo que la variable NumDormitorios es mucho más fiable a la hora de averiguar el precio del inmueble
#El modelo ideal es más complejo, pruebo con varias opciones
model3<-lm(barrio_sol,formula=Precio~NumDormitorios+NumBanyos)
summary(model3)
```

```r
model4<-lm(barrio_sol,formula=Precio~Metros.Cuadrados+NumDormitorios)
summary(model4)
```

```r
model5<-lm(barrio_sol,formula=Precio~Metros.Cuadrados+NumBanyos+NumDormitorios)
summary(model5)
```
